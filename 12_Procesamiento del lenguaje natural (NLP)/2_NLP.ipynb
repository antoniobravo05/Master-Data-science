```
from nltk.corpus import wordnet as wn
```

```
print(wn.synsets("sword"))
```

```
#lo anterior nos imprime el identificador del synset. Tenemos que mirar los métodos asociados
```

```
espada=wn.synsets("sword")[0]
#nos guardamos en espada el primer elemento de la lista para poder hacer el print directameente
```

```
print(espada.lemma_names())
```

```
#Vemos como lo anterior son sinónimos de estada en ingles
```

```
print(espada.definition())
```

```
for element in espada.hypernyms():
  print(element.lemma_names())
  print(element.definition())
```

```
for path in espada.hypernym_paths():
  for element in path:
    print(element.lemma_names())
    print(element.definition())
```

```
carruaje=wn.synsets("carriage")[0]
```

```
for path in carruaje.hypernym_paths():
  for element in path:
    print(element.lemma_names())
    print(element.definition())
```

```
#vamos como un carruaje, una misma palabra, puede entenderse desde 2 puntos distintos. Tiene 2 entidades, puede significar algo para moverse y luego un objeto físico.
```

```
#Volvemos a sward, vemos los hipónimos, es decir, los tipos de espada:
for element in espada.hyponyms():
  print(element.lemma_names())
  print(element.definition())
```

```
#Merónimos: palabras cuyo significado constituye una parte del significado total de otras palabras, denominadas éstas holónimos.
for element in espada.part_meronyms():
  print(element.lemma_names())
  print(element.definition())
```

```
import nltk
```

```
#definicion de una grrmática independiente del texto, esto puede definirse como nosotros queramos, este es un ejemplo de la profesora.

gram="""

S -> NP VP
NP -> Det N | Det N PP | 'I' 
VP -> V NP | VP PP
PP -> P NP
Det -> 'an' | 'my'
N -> 'elephant' | 'pajamas' 
V -> 'shot'
P -> 'in'

"""
```

```

gramatica_independiente_del_texto=nltk.CFG.fromstring(gram) #context free grammar, es el metodo que entiende esto.
```

```
frase="I shot an elephant in my pajamas".split()
frase
```

```
#creando el analizador
analizador=nltk.ChartParser(gramatica_independiente_del_texto)
```

```
#analizando la frase según la gramática
for analisis in analizador.parse_all(frase):
  print(analisis)
```
