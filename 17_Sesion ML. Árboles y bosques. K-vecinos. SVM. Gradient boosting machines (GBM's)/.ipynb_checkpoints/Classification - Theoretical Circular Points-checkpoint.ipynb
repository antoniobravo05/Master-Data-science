```python
import pandas as pd
%pylab inline
```

```python
df = pd.read_csv("data/points.csv")
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df["X"],df["Y"],c=df["target"])
plt.show()
```

```python
X = df[["X","Y"]].values
y = df["target"]
```

```python
from sklearn.svm import SVC
from sklearn.cross_validation import cross_val_score
```

```python
clfs=SVC(kernel="linear",C=100)
```

```python
cross_val_score(clfs,X,y,scoring="accuracy",cv=10).mean()
```

```python
clfs.fit(X,y)
```

```python

def draw(clf,X,y,h=0.05):
    
    plt.figure(figsize=(10,10))

    x_min, x_max = X[:, 0].min() - 0.05, X[:, 0].max() + .05
    y_min, y_max = X[:, 1].min() - .05, X[:, 1].max() + .05
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    if clf is not None:
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        plt.scatter(xx,yy,c=Z)
    else:
        plt.scatter(xx,yy)
    
    plt.scatter(X[:,0],X[:,1],c=y,cmap="Paired")
```

```python
draw(clfs,X,y,0.01)
```

```python
#como vemos, podemos usar SVM pero la interpretacion es que no está sirviendo. Tengo un problema, los datos no puedo separarlos con una resta, a priori.
#pero si podemos hacer una transformacion de los datos para que sean separables con una recta. Introduciendo una tercera variable, que me dará una tercera dimension
#df['Z']=df["X"]**2+df['Y']**2
#hacer esto es computacionalmente muy caro, por ello se usa el truco del kernel (ver apuntes). El kernel modifica la forma de la línea de separacion: kernel lineal:línea kernel radial:círculos, kernel polinomial:separacion con poligonos
```

```python
#Usando un kernel radial, con un parámetro adicional que es el ganmma (inversamente proporconal al radio del circulo)
clfs2=SVC(kernel="rbf",C=100,gamma=100)
cross_val_score(clfs2,X,y,scoring="accuracy",cv=10).mean()
```

```python
clfs2.fit(X,y)
```

```python
draw(clfs2,X,y,0.01)
```

```python
#Vemos que es ahora, el límite se ajusta mucho mejor a nuestros datos.
#Se está modificando la forma del espacio, en vez de tener rectas, se tienen círculos.
```

```python
from ipywidgets import interact
```

```python
@interact (gamma=[10,100,1000,10000])
def show(gamma):
    clfs2=SVC(kernel="rbf",C=100,gamma=gamma)
    clfs2.fit(X,y)
    draw(clfs2,X,y,0.01)
```

```python
#para gamma altos, overfitting
```

```python
#Buscando el ganmma optimo
a=[]
for j in range(10,1000):
    clfs2=SVC(kernel="rbf",C=100,gamma=j)
    a.append(cross_val_score(clfs2,X,y,scoring="accuracy",cv=10).mean())
```

```python
plt.plot(a)
```
