Profesor: Juan Duque.

Una recomendaión se basa en 3 elementos principales.

- Usuarios
- Productos (designados como items). Son los productos que se muestran al usuario.
- Intereacción Usuario-productos (lo que se conoce cmo rating):
    -Rating explícitos (usuario valora su interaccion con el producto). Tenemos una tupla de usuario-item-rating.
    -Rating implícitos (se establece únicamente atendiendo al comportamiento del usuario, no sabemos nada más).

Concepto de Esparsicidad: las parejas usuario-item son muy limitadas.
Curva de LongTail. Representa la frecuencia de consumo frente a los productos. Normalmente, existen pocos productos que se consumen mucho y muchos productos que se consumen poco.

¿Qué es mejor, rating implicito o explicito? Siempre es ambigüo.

¿En qué consiste en hacer una recomendación? Tenemos que estimar preferencias y luego orndar esas preferencias. Pasos:
1. Estimar preferencias: $$\hat{p}_{ui}$$
2. Ordenar preferencias estimadas: $$\hat{p}_{ui 1},\hat{p}_{ui 2},...\hat{p}_{ui n}$$

Modelos principales en un SR
- Basados en popularidad (estima las preferencias en función de las frecuencias de consumo. Son sencillos y los veremos hoy.)
- De filtrado colaborativo (son un poco mas avanzados, basados en la nocion de "similaridad" entre dos individuos o items, asociado a su vez a un concepto de "distancia" y de "representación" refiriendose a como proyectar el dato para encontrar esas distancias de forma sencilla. 
- Basados en contenido (como se establede la recomendacion atendiendo a las caracteristicas de los items)

Métricas, es como defino o cuantifico lo bien que he hecho una recomendación:
- Metricas de "Accuracy" (distancia media RMSE. Tiene mucho sentido cuando trabajamos con rating explicitos. No es la mas usada).
- Métricas de relevancia (distinguimos si pesa o no el orden de la lista que tu le das: Recall@k, precision@k. Hay un tipo de estas que tiene en cuenta el orden de las recomendaciones (se penaliza si la recomendación está lejos de los datos de test).
- Otras que no veremos tienen mas en cuenta el raking.

```python
#generar una funcon que se llame return item id que tome un texto, el diccionario y que devuelva una lista "out" de pelis que contienen el texto
def returnItemId(text,ids):
    #inicializo out e index
    out=[]
    index=[]
    #transformar todo en minusculas x.lower()
    text_lower=text.lower()
    #encontrar y.find('texto'), devuelve 0 cuando lo encuentra y un -1 cuando no
    #Recorrer todo el diccionario haciendo find para ver si lo encuentra, si es que si, lo meto en out si  no no.
    for i,j in list(item_dict.items()):
        if j.lower().find(text_lower) >= 0:
            index.append(i)
    
    for i in index:
        out.append((i,ids[i]))
        
    return out
```

### MÉTRICAS
Ver precision an recall en wikipedia.
No se puede meedir algo que el usuario no ha visto. No se puede hacer el rating si el usuario no la ha visto.
Este tipo de métricas se han adaptado un poco para aplicarlas en SR.
- recall@n (precision n, corto en n elementos), me suma la relevancia de cada item y lo normalizas por el numero de elementos en test.



```python
#implementamos el recall
#esta funcion es por usuario.
def recall_at_n(N,test,recommended):
    #solo me interesan las N primeras recomendaciones
    rec=recommended[:N]
    #inicializo
    recall=0;
    for i in rec:
        #veo si vive en test
        if i in test:
            recall+=1
    #calculo
    recall=recall/np.min([N,len(test)])#calcula el minimo dado un array
    return recall
```
